Klasa (class) to ogólny projekt, schemat lub wspomniany przepis. Mówi ona komputerowi: "Każdy samochód (Car) będzie miał markę, model i kolor".

Obiekt to konkretny egzemplarz stworzony na podstawie tego schematu. Twoje car1 (Toyota) i my_car (Honda) to właśnie obiekty.

def __init__(self, make, model, color):
To jest tzw. konstruktor. Uruchamia się automatycznie, gdy tworzysz nowy obiekt. Słowo self to sposób, w jaki obiekt mówi o samym sobie (np. "mój kolor to czerwony"). 

self.owner_number = 0
To jest atrybut (cecha). Każdy nowy samochód zaczyna z zerową liczbą właścicieli.

Metody (car_info, sell)
To są "umiejętności" obiektu.

car_info wypisuje dane o samochodzie.

sell zwiększa liczbę właścicieli o 1 przy każdej sprzedaży.
for i in range(5):
    my_car.sell()
    Skoro my_car na początku miał owner_number = 0, a pętla wywołała funkcję sell() dokładnie 5 razy, to jak myślisz, jaką liczbę właścicieli wypisze my_car.car_info() na samym końcu?

```python
class Car(object):
    def __init__(self,make,model,color):
        self.make=make;
        self.model=model;
        self.color=color;
        self.owner_number=0 
    def car_info(self):
        print("make: ",self.make)
        print("model:", self.model)
        print("color:",self.color)
        print("number of owners:",self.owner_number)
    def sell(self):
        self.owner_number=self.owner_number+1
    car1 = Car("Toyota","Corolla","red")
    my_car = Car("Honda","Civic","blue")
    car1.car_info()
    for i in range(5):
        my_car.sell()

    my_car.car_info()
```

```python
class Car(object):
    def __init__(self, make, model, color):
        self.make = make
        self.model = model
        self.color = color
        self.owner_number = 0

    def car_info(self):
        print("make: ", self.make)
        print("model:", self.model)
        print("color:", self.color)
        print("number of owners:", self.owner_number)

    def sell(self, number_of_new_owners):
        self.owner_number = self.owner_number + number_of_new_owners

# Tutaj tworzymy obiekty (już bez wcięcia, bo to główny program)
car1 = Car("Toyota", "Corolla", "red")
my_car = Car("Honda", "Civic", "blue")

car1.car_info()
my_car.sell(5)
my_car.car_info()
```
