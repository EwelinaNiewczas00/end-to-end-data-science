Zadanie 13 – Detekcja anomalii prostą metodą
Wykryj wartości odstające używając reguły 3-sigma.
Dane (5 pomiarów ciśnienia krwi):
120, 125, 118, 190, 122 mmHg
Wymagania:
Oblicz średnią i odchylenie standardowe (ręcznie, bez bibliotek)
Zidentyfikuj wartości, które odchylają się o więcej niż 3×std od średniej
Oznacz je jako anomalie
Wyświetl raport z wszystkimi wartościami i oznaczeniem anomalii
Oczekiwany rezultat:
Wartość: 120 - OK
Wartość: 125 - OK
Wartość: 118 - OK
Wartość: 190 - ANOMALIA!
Wartość: 122 - OK

```python
dane = [120, 125, 118, 190, 122]


srednia = sum(dane) / len(dane)

wariancja = sum((x-srednia)**2 for x in dane) / len(dane)
wynik = wariancja **0.5

prog_1 = srednia - 3*wynik
prog_2 = srednia + 3* wynik

print(f"Średnia: {srednia}")
print(f"Odchylenie standardowe: {wynik:.2f}")
print("-"*60)

for wartosc in dane:
    if wartosc < prog_1 or wartosc > prog_2:
        status = "Anomalia"
    else:
        status = "OK"
    print(f"Wartość: {wartosc}: {status}")
```

🧠 Zadanie 15 – Standaryzacja Z-score
Standaryzuj wartości używając wzoru z-score.
Dane (wagi pacjentów w kg):
65.5, 78.2, 54.8, 92.1, 70.0
Wymagania:
Oblicz średnią i odchylenie standardowe
Dla każdej wartości oblicz z-score: z = (x - mean) / std
Sprawdź, że średnia z-scores ≈ 0 i odchylenie ≈ 1
Wyświetl oryginalną wartość i jej z-score
Zinterpretuj: z > 2 (wartość bardzo wysoka), |z| < 1 (typowa), z < -2 (bardzo niska)

```python
import statistics
import numpy as np

wagi = [65.5, 78.2, 54.8, 92.1, 70.0]
srednia = statistics.mean(wagi)
print(srednia)
srednia_np = np.mean(wagi)
print(srednia_np)
odchylenie = statistics.pstdev(wagi)
print(odchylenie)
```

```python
wagi = [65.5, 78.2, 54.8, 92.1, 70.0]
n = len(wagi)

suma = 0
for w in wagi:
    suma+=w
srednia = suma/n

suma_kwadratów = 0
for w in wagi:
    suma_kwadratów += (w-srednia) **2
std = (suma_kwadratów/n)**0.5

z_scores = []

print(f"{'Waga [kg]':<10} | {'Z-score':<10} | {'Interpretacja'}")
print("-" * 45)

for w in wagi:
    z= (w-srednia)/std
    z_scores.append(z)

    if z >2:
        opis = "Bardzo wysoka"
    elif z<-2:
        opis = "Bardzo niska"
    elif abs(z) <1:
        opis = "Typowa"
    else:
        opis = "Poza normą"

    print(f"{w:<10.1f} | {z:<10.2f} | {opis}")

srednia_z = sum(z_scores) / len(z_scores)
std_z = (sum((z - srednia_z)**2 for z in z_scores) / len(z_scores)) ** 0.5


print("-" * 45)
print(f"Sprawdzenie: Średnia Z-score = {srednia_z:.1f}, Odchylenie Z-score = {std_z:.1f}")



```
