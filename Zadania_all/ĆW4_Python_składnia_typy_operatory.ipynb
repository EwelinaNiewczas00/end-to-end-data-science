Zadanie 1 – Kalkulator BMI
Napisz program, który obliczy wskaźnik BMI (Body Mass Index).
Dane:
Waga: 70 kg
Wzrost: 1.75 m
Wymagania:
Stwórz zmienne waga i wzrost
Oblicz BMI według wzoru: BMI = waga / (wzrost²)
Wyświetl wynik z 2 miejscami po przecinku
Dodaj interpretację: BMI < 18.5 (niedowaga), 18.5-24.9 (norma), ≥25 (nadwaga)
Oczekiwany rezultat:
BMI: 22.86
Interpretacja: Norma

```python
#dane
waga = 70
wzrost = 1.75
#obliczenie
bmi = waga/(wzrost**2)
print(f"Przy twojej wadzę: {waga}kg i podanym wzroście: {wzrost}m - BMI wynosi: {bmi:.2f}")

if bmi < 18.5:
    print(f"Twoje BMI: {bmi:.2f} stwierdza niedowagę")
elif 18.5 <bmi < 24.9:
    print(f"Twoje BMI: {bmi:.2f} stwierdza normę")
else:
    print(f"Twoje BMI: {bmi:.2f} stwierdza nadwagę")


```

Zadanie 2 – Konwersja temperatur
Napisz program konwertujący temperatury Celsius na Fahrenheit i Kelvin.
Dane:
Temperatura: 23.5°C
Wymagania:
Fahrenheit = Celsius × 9/5 + 32
Kelvin = Celsius + 273.15
Wyświetl wszystkie 3 wartości sformatowane z 1 miejscem po przecinku
Oczekiwany rezultat:
23.5°C = 74.3°F = 296.6 K

```python
temp = 23.5
fahr = temp *(9/5) +32
kelv = temp + 273.15

print(f"{temp:.1f}°C = {fahr:.1f}°F = {kelv:.1f}K")
```

Zadanie 3 – Podział zbioru danych
Oblicz rozmiary zbiorów treningowego, walidacyjnego i testowego.
Dane:
Całkowita liczba próbek: 1000
Proporcje: 70% trening, 15% walidacja, 15% test
Wymagania:
Użyj mnożenia i dzielenia całkowitego
Upewnij się, że suma = 1000 (obsłuż resztę z dzielenia)
Wyświetl rozmiary wszystkich zbiorów

```python
próby = 1000

trening = (próby * 70) //100
walidacja = (próby * 15) //100
test = próby - (trening+walidacja)
wynik_spr = trening + walidacja + test

print(f"Rozmiar próby: {próby}\n Trenig: {trening}\n Walidacja: {walidacja}\n Test: {test}\n ")
print(wynik_spr)
```
