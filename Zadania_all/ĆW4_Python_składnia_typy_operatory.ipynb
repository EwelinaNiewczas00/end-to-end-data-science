Zadanie 1 – Kalkulator BMI
Napisz program, który obliczy wskaźnik BMI (Body Mass Index).
Dane:
Waga: 70 kg
Wzrost: 1.75 m
Wymagania:
Stwórz zmienne waga i wzrost
Oblicz BMI według wzoru: BMI = waga / (wzrost²)
Wyświetl wynik z 2 miejscami po przecinku
Dodaj interpretację: BMI < 18.5 (niedowaga), 18.5-24.9 (norma), ≥25 (nadwaga)
Oczekiwany rezultat:
BMI: 22.86
Interpretacja: Norma

```python
# dane
waga = 70
wzrost = 1.75
# obliczenie
bmi = waga / (wzrost**2)
print(
    f"Przy twojej wadzę: {waga}kg i podanym wzroście: {wzrost}m - BMI wynosi: {bmi:.2f}"
)

if bmi < 18.5:
    print(f"Twoje BMI: {bmi:.2f} stwierdza niedowagę")
elif 18.5 < bmi < 24.9:
    print(f"Twoje BMI: {bmi:.2f} stwierdza normę")
else:
    print(f"Twoje BMI: {bmi:.2f} stwierdza nadwagę")
```

 Zadanie 7 – Porównanie dwóch modeli
Określ, który model jest lepszy.
Dane:
Model A: accuracy = 0.87, train time = 120s
Model B: accuracy = 0.89, train time = 450s
Wymagania:
Porównaj accuracy (różnica > 0.02 jest znacząca)
Porównaj czas treningu
Użyj operatorów logicznych do określenia "najlepszego kompromisu"
Wyświetl rekomendację

```python
accuracy_a = 0.87
train_time_a = 120

accuracy_b = 0.89
train_time_b = 450

roznica_acc = round(accuracy_b - accuracy_a, 2)

if roznica_acc > 0.02 and train_time_b < train_time_a:
    print("Model B jest lepszy i szybszy. Rekomendacja: Model B")
elif roznica_acc > 0.02 and train_time_b > train_time_a:
    print(
        "Model B jest lepszy, ale wolniejszy. Opłaca się czekać. Rekomendacja: Model B"
    )
elif roznica_acc <= 0.02 and train_time_a < train_time_b:
    print(f"Różnica w dokładności ({roznica_acc}) nie jest znacząca.")
    print("Model A jest znacznie szybszy. Najlepszy: Model A")
else:
    print("Modele są bardzo podobne. Rekomendacja: Obojętnie")
```

9. ✏ Zadanie 9 – Normalizacja Min-Max
Znormalizuj wartość do zakresu [0, 1].
Dane:
Wartość: 75
Minimum w kolumnie: 50
Maximum w kolumnie: 100
Wymagania:
Wzór: normalized = (value - min) / (max - min)
Oblicz dla wartości 75, 50, 100, 62.5
Sprawdź, czy wynik mieści się w [0, 1]
Wyświetl tabelkę: wartość oryginalna | wartość znormalizowana

```python
wartości = [75, 50, 100, 62.5]
min_kol = 50
max_kol = 100
print(f"{"Wartość oryginalna":<20}  {"Wartość znormalizowana":<20}  {"Status":<20}")
print("--" * 40)
for wartość in wartości:
    normalized = (wartość - min_kol) / (max_kol - min_kol)
    if 0 <= normalized <= 1:
        status = "Mieści się w przedziale 0-1"
    else:
        status = "Nie mieści się w przedziale 0-1"

    print(f"{wartość:<20} | {normalized:<20} | {status:<20}")
print()
# list comprehension
znormalizowane = [(x - min_kol) / (max_kol - min_kol) for x in wartości]
print(znormalizowane)
```

13. 🧠 Zadanie 13 – Detekcja anomalii prostą metodą
Wykryj wartości odstające używając reguły 3-sigma.
Dane (5 pomiarów ciśnienia krwi):
120, 125, 118, 190, 122 mmHg
Wymagania:
Oblicz średnią i odchylenie standardowe (ręcznie, bez bibliotek)
Zidentyfikuj wartości, które odchylają się o więcej niż 3×std od średniej
Oznacz je jako anomalie
Wyświetl raport z wszystkimi wartościami i oznaczeniem anomalii
Oczekiwany rezultat:

```python
pomiary = [120, 125, 118, 190, 122]
srednia = sum(pomiary) / len(pomiary)
print(f"Średnia : {srednia}")

licznik_razem = 0

for pomiar in pomiary:
    licznik = (pomiar - srednia) ** 2
    licznik_razem += licznik
odchylenie = (licznik_razem / len(pomiary)) ** 0.5


for pomiar in pomiary:
    if abs(pomiar - srednia) > 3 * odchylenie:
        print(f"Wartość: {pomiar} - ANOMALIA")
    else:
        print(f"Wartość: {pomiar} - OK")

print(f"Odchylenie standardowe: {odchylenie:.2f}")


print()
print()
# Rozwiązanie chat:
import numpy as np

pomiary = [120, 125, 118, 190, 122]

# Ta jedna komenda pod spodem robi dokładnie to, co Twoja pętla! (std = standard deviation)
odchylenie = np.std(pomiary)

print(f"Odchylenie (NumPy): {odchylenie}")
```

20. 🧠 Zadanie 20 – Generowanie SQL query z Python
Wygeneruj zapytanie SQL używając zmiennych Python.
Dane:
Tabela: "patients"
Kolumny do wybrania: ["patient_id", "age", "diagnosis"]
Warunek 1: age > 40
Warunek 2: diagnosis w ['diabetes', 'hypertension']
Sortowanie: według age malejąco
Limit: 100 wierszy
Wymagania:
Użyj f-strings do budowy zapytania SELECT
Poprawne formatowanie SQL (wielkie litery dla słów kluczowych)
Lista diagnoz jako string: "('diabetes', 'hypertension')"
Użyj join() do łączenia kolumn
Wyświetl gotowe zapytanie, które można skopiować do SQL
Oczekiwany format:
SELECT patient_id, age, diagnosis
FROM patients
WHERE age > 40 AND diagnosis IN ('diabetes', 'hypertension')
ORDER BY age DESC
LIMIT 100;

```python
# 1. Nasze dane zebrane w zmiennych (w prawdziwym życiu pobralibyśmy je np. z aplikacji)
tabela = "patients"
kolumny = ["patient_id", "age", "diagnosis"]
wiek_min = 40
diagnozy = ["diabetes", "hypertension"]
limit = 100

# 2. Przygotowanie danych do wstawienia (Magia Pythona)
# Łączymy nazwy kolumn przecinkiem:
kolumny_str = ", ".join(kolumny)

# Tworzymy listę diagnoz z apostrofami, a potem łączymy w stringa w nawiasach
# Używamy sprytnego wyrażenia listowego (List Comprehension), o którym mówiliśmy wcześniej!
diagnozy_str = ", ".join([f"'{d}'" for d in diagnozy])
diagnozy_sql = f"({diagnozy_str})"

# 3. Budujemy zapytanie używając wielolinijkowego f-stringa (trzy cudzysłowy """)
query = f"""
SELECT {kolumny_str}
FROM {tabela}
WHERE age > {wiek_min} AND diagnosis IN {diagnozy_sql}
ORDER BY age DESC
LIMIT {limit};
"""

# Metoda .strip() po prostu usuwa puste entery na początku i końcu tekstu
print(query.strip())
```
