Napisz funkcję statystyki(liczby), która przyjmuje listę liczb i zwraca tuple zawierający:
(suma, średnia, minimum, maksimum).
Dataset:
Lista ocen: [3.5, 4.0, 5.0, 3.0, 4.5, 5.0, 4.0, 3.5]
Wymagania:
Funkcja powinna zwracać tuple z 4 wartościami
Średnia zaokrąglona do 2 miejsc po przecinku
Oczekiwany rezultat:
Wypisanie wszystkich statystyk na ekran

```python
oceny = [3.5, 4.0, 5.0, 3.0, 4.5, 5.0, 4.0, 3.5]

def statystyki(x):
    suma = sum(x)
    srednia = sum(x) / len(x)
    minimum = min(x)
    maks = max(x)
    return (suma, srednia, minimum, maks)
print(statystyki(oceny))

statystyka = lambda x: (sum(x), sum(x)/len(x), min(x), max(x)) 
wyniki = statystyka(oceny)
suma, srednia, minimum, maksimum =wyniki
print(f"Suma {suma} średnia {srednia} Minimum {minimum} Maksimum {maksimum}")
```
