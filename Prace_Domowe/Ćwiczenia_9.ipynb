```python
from Prace_Domowe.Ćwiczenia_5 import nazwa


class StatDataset:
    def __init__(self, nazwa, wartości):
        self.nazwa = nazwa
        self.wartości = wartości
    def średnia(self):
        return sum(self.wartości) / len(self.wartości)
    def minimun(self):
        return min(self.wartości)
    def maximum(self):
        return max(self.wartości)
    def raport(self):
        print(f"===Raport dla {self.nazwa}===")
        print(f"Średnia: {self.średnia():.2f}")
        print(f"Min: {self.minimun()}")
        print(f"Max: {self.maximum()}")
        print(f"Zakres: {self.maximum() - self.minimun()}")
dane_sprzedaży = StatDataset("Sprzedaż miesięczna", [1200, 1500, 1350, 1800, 1600, 1700])
dane_sprzedaży.raport()
```

```python
class Model:
    def __init__(self, nazwa):
        self.nazwa = nazwa
        self.wytrenowany = False
    def trenuj(self, dane):
        print(f"Trenuję model {self.nazwa}  ")
        self.wytrenowany = True
    def czy_wytrenowany(self):
        return self.wytrenowany

class ModelRegresji(Model):
    def __init__(self, nazwa, typ_regresji="liniowa"):
        super().__init__(nazwa)
        self.typ_regresji = typ_regresji

    def przewiduj(self, x):
        if not self.wytrenowany:
            print("Błąd: Model nie jest wytrenowany!")
            return None
        return 2.5*x+1.3

model_reg = ModelRegresji("Regresja Sprzedaży", typ_regresji="liniowa")
print(f"Czy wytrenowany: {model_reg.czy_wytrenowany()}")
print()
model_reg.trenuj([1, 2, 3, 4, 5]) # Metoda z klasy bazowej
print(f"Czy wytrenowany: {model_reg.czy_wytrenowany()}")
print()
predykcja = model_reg.przewiduj(10) # Metoda specyficzna dla ModelRegresji
print(f"Predykcja dla x=10: {predykcja}")
```

```python
class Transformer:
    def __init__(self, nazwa):
        self.nazwa = nazwa
    def transformuj(self, dane):
        raise NotImplementedError
class Normalizator(Transformer):
    def __init__(self):
        super().__init__("Normalizator")
        self.min_val =None
        self.max_val = None
    def transformuj(self, dane):
        self.min_val = min(dane)
        self.max_val = max(dane)
        zakres = self.max_val - self.min_val

        if zakres == 0:
            return [0.5]*len(dane)
        return [(x-self.min_val) / zakres for x in dane]
class Standaryzator(Transformer):
    def __init__(self):
        super().__init__("Standaryzator")
        self.srednia = None
        self.odch = None

    def transformuj(self, dane):
        self.srednia = sum(dane) / len(dane)
        wariancja = sum((x-self.srednia) **2 for x in dane) / len(dane)
        self.odch = wariancja ** 0.5
        if self.odch ==0:
            return[0,0] * len(dane)
        return[(x-self.srednia) / self.odch for x in dane]


dane = [10,20,30,40,50]


norm = Normalizator()
dane_norm = norm.transformuj(dane)
print(f"Dane org: {dane}")
print(f"Po normalizacji: {[round(x, 2) for x in dane_norm]}")
print(f"min: {norm.min_val}, max: {norm.max_val}")

stand = Standaryzator()
dane_stand = stand.transformuj(dane)
print(f"Po standaryzacji: {[round(x,2) for x in dane_stand]}")
print(f"Średnia: {stand.srednia}, Odchylenie: {round(stand.odch, 2)}")
```

```python
class Raport:
    def __init__(self, tytuł):
        self.tytuł = tytuł
    def generuj(self):
        print(f"==={self.tytuł}===")
        print(f"Brak szczegółowych danych")
class RaportSczegółowy(Raport):
    def __init__(self, tytuł, dane):
        super().__init__(tytuł)
        self.dane = dane
    def generuj(self):
        print(f"==={self.tytuł}===")
        print(f"Liczba obserwacji: {len(self.dane)}")
        print(f"Średnia {sum(self.dane)/len(self.dane):.2f}")
        print(f"Min: {min(self.dane)}, Max: {max(self.dane)}")

raport1 = Raport("Raport podtawowy")
raport1.generuj()

raport2 = RaportSczegółowy("Raport szczegółowy", [100,200,300,400,220])
raport2.generuj()
```

```python
class Czyszczenie:
    def __init__(self):
        self.nazwa = "Czyszczenie danych"

    def wykonaj(self, dane):
        print(f" [{self.nazwa}]")
        return [x for x in dane if x >=0]
class Normalizacja:
    def __init__(self):
        self.nazwa ="Normalizacja"
    def wykonaj(self, dane):
        print(f" [{self.nazwa}]")
        if len(dane) == 0:
            return[]
        min_val = min(dane)
        max_val = max(dane)
        zakres = max_val - min_val
        if zakres == 0:
            return [0.5] * len(dane)
        return[(x-min_val)/zakres for x in dane]
class Agregacja:
    def __init__(self):
        self.nazwa = "Agregacja"
    def wykonaj(self, dane):
        print(f" [{self.nazwa}]")
        if len(dane) ==0:
            return 0
        return sum(dane) / len(dane)

class Pipeline:
    def __init__(self, nazwa):
        self.nazwa = nazwa
        self.kroki =[]
    def dodaj_krok(self, krok):
        self.kroki.append(krok)
    def uruchom(self, dane):
        print(f"pipeline: {self.nazwa}")
        wynik = dane
        for krok in self.kroki:
            wynik = krok.wykonaj(wynik)
        print(f"Wynik końcowy: {wynik}")
        return wynik


pipeline = Pipeline("Przetwarzanie danych sprzedażowych")
pipeline.dodaj_krok(Czyszczenie())
pipeline.dodaj_krok(Normalizacja())
pipeline.dodaj_krok(Agregacja())
dane_wej = [10, -5, 20, 30, -2, 40, 50]
wynik = pipeline.uruchom(dane_wej)
```

```python
class Produkt:
    def __init__(self, nazwa, cena, ilość_w_magazynie):
        self.nazwa = nazwa
        self.cena = cena
        self.ilość_w_magazynie = ilość_w_magazynie

    def pokaz_info(self):
        print(f"Nazwa produktu: {self.nazwa}, cena: {self.cena}, ilość w magazynie: {self.ilość_w_magazynie}")
    def sprzedaj(self, ilość):
        self.ilość_w_magazynie -= ilość
        print(f"Sprzedano {ilość}.Na magazynie powinno być {self.ilość_w_magazynie} produktów")
        if self.ilość_w_magazynie < 0:
            print ("Niedobór na magazynie")
        else:
            print("Wszystko w porządku")
    def dostawa(self, ilość):
        self.ilość_w_magazynie += ilość
        print(f"Po dostawie na magazynie znajduję się {self.ilość_w_magazynie} produktów")

produkty = Produkt("laptop", 5000, 10)
produkty.pokaz_info()
produkty.sprzedaj(3)
produkty.dostawa(6)




```

✏ Zadanie 2 – Klasa Kalkulator
Stwórz klasę Kalkulator z metodami: dodaj(a, b), odejmij(a, b), pomnoz(a, b),
podziel(a, b).
Stwórz obiekt i wykonaj przykładowe obliczenia.

```python
class Kalkulator:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def dodaj(self):
        print(f"Wynik dodawania: {self.a + self.b}")
    def odejmij(self):
        print(f"Wynik odejmowania: {self.a - self.b}")
    def pomnoz(self):
        print(f"Wynik mnożenia: {self.a * self.b}")
    def podziel(self):
        print(f"Wynik dzielenia: {self.a / self.b}")

oblicz = Kalkulator(10,5)
oblicz.dodaj()
oblicz.odejmij()
oblicz.pomnoz()
oblicz.podziel()
```
