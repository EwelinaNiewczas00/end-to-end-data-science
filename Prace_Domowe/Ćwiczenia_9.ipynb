```python
class Dataset:

    def __init__(self, nazwa):

        self.nazwa = nazwa # Atrybut (cecha) obiektu
        self.dane = [] # Lista na dane
    def dodaj_wiersz(self, wiersz):

        self.dane.append(wiersz)
    def pokaz_info(self):

        print(f"Dataset: {self.nazwa}")
        print(f"Liczba wierszy: {len(self.dane)}")
# Tworzenie obiektów (instancji klasy)
ds1 = Dataset("Sprzedaż Q1")
ds1.dodaj_wiersz([100, 200, 300])
ds1.dodaj_wiersz([150, 250, 350])
ds1.pokaz_info()
# Drugi obiekt tej samej klasy
ds2 = Dataset("Sprzedaż Q2")
ds2.dodaj_wiersz([180, 220, 320])
ds2.pokaz_info()
```
