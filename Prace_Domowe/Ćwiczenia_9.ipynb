```python
class Model:
    def __init__(self, nazwa):
        self.nazwa = nazwa
        self.wytrenowany = False
    def trenuj(self, dane):
        print(f"Trenuję model {self.nazwa}  ")
        self.wytrenowany = True
    def czy_wytrenowany(self):
        return self.wytrenowany

class ModelRegresji(Model):
    def __init__(self, nazwa, typ_regresji="liniowa"):
        super().__init__(nazwa)
        self.typ_regresji = typ_regresji

    def przewiduj(self, x):
        if not self.wytrenowany:
            print("Błąd: Model nie jest wytrenowany!")
            return None
        return 2.5*x+1.3

model_reg = ModelRegresji("Regresja Sprzedaży", typ_regresji="liniowa")
print(f"Czy wytrenowany: {model_reg.czy_wytrenowany()}")
print()
model_reg.trenuj([1, 2, 3, 4, 5]) # Metoda z klasy bazowej
print(f"Czy wytrenowany: {model_reg.czy_wytrenowany()}")
print()
predykcja = model_reg.przewiduj(10) # Metoda specyficzna dla ModelRegresji
print(f"Predykcja dla x=10: {predykcja}")
```

```python
class Transformer:
    def __init__(self, nazwa):
        self.nazwa = nazwa
    def transformuj(self, dane):
        raise NotImplementedError
class Normalizator(Transformer):
    def __init__(self):
        super().__init__("Normalizator")
        self.min_val =None
        self.max_val = None
    def transformuj(self, dane):
        self.min_val = min(dane)
        self.max_val = max(dane)
        zakres = self.max_val - self.min_val

        if zakres == 0:
            return [0.5]*len(dane)
        return [(x-self.min_val) / zakres for x in dane]
class Standaryzator(Transformer):
    def __init__(self):
        super().__init__("Standaryzator")
        self.srednia = None
        self.odch = None

    def transformuj(self, dane):
        self.srednia = sum(dane) / len(dane)
        wariancja = sum((x-self.srednia) **2 for x in dane) / len(dane)
        self.odch = wariancja ** 0.5
        if self.odch ==0:
            return[0,0] * len(dane)
        return[(x-self.srednia) / self.odch for x in dane]


dane = [10,20,30,40,50]


norm = Normalizator()
dane_norm = norm.transformuj(dane)
print(f"Dane org: {dane}")
print(f"Po normalizacji: {[round(x, 2) for x in dane_norm]}")
print(f"min: {norm.min_val}, max: {norm.max_val}")

stand = Standaryzator()
dane_stand = stand.transformuj(dane)
print(f"Po standaryzacji: {[round(x,2) for x in dane_stand]}")
print(f"Średnia: {stand.srednia}, Odchylenie: {round(stand.odch, 2)}")
```

```python
class Czyszczenie:
    def __init__(self):
        self.nazwa = "Czyszczenie danych"

    def wykonaj(self, dane):
        print(f" [{self.nazwa}]")
        return [x for x in dane if x >=0]
class Normalizacja:
    def __init__(self):
        self.nazwa ="Normalizacja"
    def wykonaj(self, dane):
        print(f" [{self.nazwa}]")
        if len(dane) == 0:
            return[]
        min_val = min(dane)
        max_val = max(dane)
        zakres = max_val - min_val
        if zakres == 0:
            return [0.5] * len(dane)
        return[(x-min_val)/zakres for x in dane]
class Agregacja:
    def __init__(self):
        self.nazwa = "Agregacja"
    def wykonaj(self, dane):
        print(f" [{self.nazwa}]")
        if len(dane) ==0:
            return 0
        return sum(dane) / len(dane)

class Pipeline:
    def __init__(self, nazwa):
        self.nazwa = nazwa
        self.kroki =[]
    def dodaj_krok(self, krok):
        self.kroki.append(krok)
    def uruchom(self, dane):
        print(f"pipeline: {self.nazwa}")
        wynik = dane
        for krok in self.kroki:
            wynik = krok.wykonaj(wynik)
        print(f"Wynik końcowy: {wynik}")
        return wynik


pipeline = Pipeline("Przetwarzanie danych sprzedażowych")
pipeline.dodaj_krok(Czyszczenie())
pipeline.dodaj_krok(Normalizacja())
pipeline.dodaj_krok(Agregacja())
dane_wej = [10, -5, 20, 30, -2, 40, 50]
wynik = pipeline.uruchom(dane_wej)
```

```python
class Pipeline:
    def __init__(self, dane):
        self.dane = dane
    def filtrowanie_dodatnich(self):
        self.dane = [dana for dana in self.dane if dana>0]
        print(f" dane dodatnie: {self.dane}")
        return self
    def zaokraglenie(self):
        self.dane = [round(dana) for dana in self.dane]
        print(f" Zaokrąglenie dodatnich do liczb całkowutych: {self.dane}")
        return self
    def skalowanie(self):
        self.dane = [dana *0.5 for dana in self.dane]
        return self
    def sumowanie(self):
        suma = sum(self.dane)
        print(f"Suma: {suma}")
        return suma
    def cały_proces(self):
        return self.filtrowanie_dodatnich().zaokraglenie().skalowanie().sumowanie()
liczby = [1.7, -3.2, 5.8, -1.1, 8.3, 2.4]
pipeline = Pipeline(liczby)
wynik = pipeline.filtrowanie_dodatnich().zaokraglenie().skalowanie().sumowanie()
wynik2 =pipeline.cały_proces()
wynik3 = pipeline.sumowanie()
```

10. ✏ Zadanie 10 – Klasa Zamowienie z kompozycją
Stwórz klasę Produkt (nazwa, cena) oraz klasę Zamowienie , która zawiera listę
produktów.
Metody Zamowienie :
Stwórz zamówienie z 5 produktami.
(średnie)
dodaj_produkt(produkt, ilosc)
laczna_wartosc() – suma wartości zamówienia
pokaz_zamowienie() – wyświetla wszystkie produkty i łączną wartość


```python
class Produkt():
    def __init__(self, nazwa, cena):
        self.nazwa= nazwa
        self.cena = cena
    def __str__(self):
        return self.nazwa

class Zamówienie():
    def __init__(self):
        self.lista_produktów =[]

    def dodaj_produkt(self, produkt : Produkt , ilość):
        self.lista_produktów.append((produkt, ilość))
        print(f"Do koszyka dodałes {produkt} w ilości{ilość} sztuk")
        return self
    def łączna_wartość(self, suma=0):
        for produkt, ilość in self.lista_produktów:
            suma = suma + (produkt.cena *ilość)
        print(f"Suma wynos: {suma}")
        return suma
    def pokaż_zamówienie(self):
        for produkt, ilość in self.lista_produktów:
            print(f"- {produkt.nazwa}: {ilość} szt.")
        print(f"za zamówienie zapłacisz {self.łączna_wartość()} zł")
        return self
    def __repr__(self):
        return f"--- paragon ---"

p1 = Produkt("jabłko", 4.3)
p2 = Produkt("banan", 7.9)
p3 = Produkt("gruszka", 8.8)
p4 = Produkt("mandarynka", 5.4)
p5 = Produkt("pomarańcz", 2.3)
moje_zakupy = Zamówienie()

moje_zakupy.dodaj_produkt(p1, 7).dodaj_produkt(p2, 8).dodaj_produkt(p3, 5)
moje_zakupy.dodaj_produkt(p4, 2).dodaj_produkt(p5, 4)
moje_zakupy.pokaż_zamówienie()
```

```python
import math
class Figura:
    def pole(self):
        return 0
class prostokąt(Figura):
    def __init__(self, szerokość, wysokość):
        self.szerokość = szerokość
        self.wysokość = wysokość
    def pole(self):
        return self.szerokość * self.wysokość
    def __str__(self):
        return "prostokąt "
class koło(Figura):
    def __init__(self, promień):
        self.promień = promień
    def pole(self):
        return math.pi * (self.promień ** 2)
    def __str__(self):
        return "koło "
moje_figury = [
    prostokąt(8, 6),
    koło(4),
    prostokąt(2, 5)
]
suma_pól = 0
print("obliczanie pól figur:")
for figura in moje_figury:
    wynik_pola = figura.pole()
    print(f"Pole tej figury {figura} wynosi: {wynik_pola:.2f}")
    suma_pól += wynik_pola

print(f"Pole łącznie: {suma_pól:.2f}")
```
