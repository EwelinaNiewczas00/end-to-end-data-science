Zadanie 1 - Kalkulator BMI

Napisz program, który obliczy wskaźnik BMI (Body Mass Index).

Dane:

• Waga: 70 kg

• Wzrost: 1.75 ml

Wymagania:

• Stwórz zmienne waga i wzrost

• Oblicz BMI według wzoru: BMI = waga / (wzrost')

• Wyświetl wynik z 2 miejscami po przecinku

• Dodaj interpretację: BMI < 18.5 (niedowaga), 18.5-24.9 (norma), ≥25 (nadwaga)

Oczekiwany rezultat:

BMI: 22.86

Interpretacja: Norma#"""


```python
#zadanie1
waga = 70 
wzrost = 1.75 

BMI = waga / (wzrost ** 2)
print (f'dla podanej wagi {waga} kg i wzrostu {wzrost} metra BMI wynosi:' + "  " f'{BMI:.2F}'  , end=" ") 
if BMI < 18.5:
    print("interpretacja: niedowaga")
elif 18.5 <= BMI <= 24.9:
    print("interpretacja: norma")
else:
    print("interpretacja: nadwaga")
```

Zadanie 12 – Estymacja czasu treningu
Oblicz szacowany czas treningu modelu.
Dane:
Wymagania:
(średnie)
Liczba próbek: 50000
Batch size: 128
Liczba epok: 50
Czas na jeden batch: 0.15 sekundy
Oblicz liczbę batchy na epokę (użyj dzielenia całkowitego i modulo)
Oblicz całkowity czas w sekundach
Skonwertuj na minuty i godziny
Wyświetl: "Szacowany czas: X h Y min Z s

```python
#Zadanie12
liczba_próbek = 50000 # tyle probek do wsadzenia 
bath_size = 128 #jedna iteracja modelu ile danych naraz? Tyle miesci się w jednej kontenerze 
liczba_epok = 50 #jak długo trenujemy, czyli musze 50 razy liczby probek trenowac i pakowac do kontenera 
jeden_batch = 0.15 #jak szybko jeden model uczy się na jednej paczce danych, tyle na jeden kontener

jeden_kontener = liczba_próbek // bath_size
reszta = liczba_próbek % bath_size
if reszta == 0:
    kontener = jeden_kontener 
else:
    kontener = jeden_kontener +1
print(f"Na jedno przejscie potrzeba {kontener} kontenerów")

epoki = liczba_epok * kontener
print(f"Wykonamy {epoki} razy dla {liczba_epok} epok")

czas_sekundy = epoki * jeden_batch
minuty = czas_sekundy//60
sekundy = czas_sekundy % 60
print(f"Wszystko zajmię {czas_sekundy} sekund czyli przleiczając {minuty} minut i {sekundy} sekund")

```
