Zadanie 11 – Normalizacja Grupowa
Dla datasetu California Housing znormalizuj MedInc (dochód) w każdym regionie do
zakresu [0, 1].
Dataset: California Housing
Wymagania:
Stwórz regiony (np. na podstawie Latitude > median)
Napisz funkcję normalizacji Min-Max
Użyj .groupby().transform() lub .apply()
Zweryfikuj, że min=0 i max=1 w każdym regionie
Oczekiwany rezultat:
DataFrame z kolumną znormalizowaną grupowo

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

housing = fetch_california_housing(as_frame=True)
df_housing = housing.frame

mediana_lat = df_housing["Latitude"].median()
df_housing["Region"] = df_housing["Latitude"].apply(lambda x: "Polnoc" if x > mediana_lat else "Poludnie")

def iqr_func(x):
    q3 = x.quantile(0.75)
    q1 = x.quantile(0.25)
    return q3 - q1

tabela_iqr = df_housing.groupby("Region")["MedHouseVal"].agg(iqr_func).reset_index()
tabela_iqr.columns = ["Region", "IQR"]

def normalize_min_max(x):
    return (x - x.min()) / (x.max() - x.min())

df_housing["MedInc_Normalized"] = df_housing.groupby("Region")["MedInc"].transform(normalize_min_max)


print("REGIONY TABELA")
print(tabela_iqr)

print("\nNORMALIZACJA")
weryfikacja = df_housing.groupby("Region")["MedInc_Normalized"].agg(["min", "max"])
print(weryfikacja)

print("\nDATAFRAME")
print(df_housing[["Region", "MedHouseVal", "MedInc", "MedInc_Normalized"]].head())
```
