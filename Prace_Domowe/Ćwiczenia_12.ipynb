```python
import pandas as pd
import numpy as np

html = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

baza = pd.read_csv(html)

baza2 = baza.groupby(["Pclass", "Sex"])["Age"].mean().unstack()

print(baza2)

baza3 = baza.pivot_table(index =["Pclass"], columns=["Survived"], aggfunc="size")

print(baza3)
```
