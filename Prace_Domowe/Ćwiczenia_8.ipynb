```python
def oblicz_srednia(lista_liczb):
    suma = 0
    for liczba in lista_liczb:
        suma += liczba
    srednia = suma / len(lista_liczb)
    return srednia

sprzedaz = [120.5, 145.5, 98.7, 167.3, 134.6]

srednia_sprzedaz = oblicz_srednia(sprzedaz)
print(f"Średnia sprzedaż : {srednia_sprzedaz:.2f} tys. zł PLN")
```

```python
def przygotuj_dane(lista_wartości, min_wartość, max_wartość):
    czyste_dane = []
    for wartość in lista_wartości:
        if min_wartość <= wartość <= max_wartość:
            czyste_dane.append(wartość)
    return czyste_dane
    
temperatury = [22.1, 23.5, -999, 24.3, 25.1, 100, 23.8, 22.9]

temp_ok = przygotuj_dane(temperatury, 15, 35)
print(f"Oryginalne temperatury: {temperatury} liczba pomiarów {len(temperatury)}")
print(f"Temperatur pofiltrowane: {temp_ok}, liczba pomiarów: {len(temp_ok)}")




```

```python


def normalizuj_dane(dane, metoda="min-max", zaokrąglenie =2):
    if metoda =="min-max":
        min_val = min(dane)
        max_val = max(dane)
        zakres = max_val - min_val 

        znormalizowane = []
        for wartość in dane:
            norma = (wartość - min_val) / zakres
            znormalizowane.append(round(norma, zaokrąglenie))
        return znormalizowane 
    elif metoda == "z-score":
        srednia = sum(dane) /len(dane)
        suma_kw = sum((x-srednia)**2 for x in dane)
        odchylenie = (suma_kw / len(dane))**0.5

        znormalizowane = []
        for wartosc in dane:
            z= (wartosc - srednia) / odchylenie
            znormalizowane.append(round(z, zaokrąglenie))
        return znormalizowane 
ceny = [49.99, 89.99, 129.99, 199.99, 299.99]
ceny_minmax = normalizuj_dane(ceny)
print(ceny)
print(ceny_minmax)
```

```python
def statystyki_opisowe(dane):
    srednia = sum(dane) / len(dane)
    posortowane = sorted(dane)
    n = len(posortowane)
    if n % 2 ==0:
        mediana =  (posortowane[n//2 -1] + posortowane[n//2])/2
    else:
        mediana =  [posortowane[n//2]]
    
    minimum= min(dane)
    maksimum = max(dane)

    zakres = maksimum - minimum
    return srednia, mediana, minimum, maksimum, zakres
 
czas_pracy = [8.5, 7.2, 9.1, 8.0, 6.5, 10.2, 7.8, 8.3, 9.5, 7.9]
sr, med, mini, maxi, zak = statystyki_opisowe(czas_pracy)
print("Statystyki czasu pracy:")
print(f" Średnia: {sr:.1f}h")
print(f" Mediana: {med:.1f}h")
print(f" Minimum: {mini:.1f}h")
print(f" Maksimum: {maxi:.1f}h")
print(f" Zakres: {zak:.1f}h")
```

```python
# 1. TWOJA BAZA KLIENTÓW (SŁOWNIKI W LIŚCIE)
# Każdy klient to osobna "kartoteka" {}
klienci = [
    {"id": 1, "wiek": 25, "dochod": 4500, "wydatki": 3200},
    {"id": 2, "wiek": 35, "dochod": 7800, "wydatki": 5100},
    {"id": 3, "wiek": 45, "dochod": 9200, "wydatki": 4800},
    {"id": 4, "wiek": 28, "dochod": 5200, "wydatki": 4900},
    {"id": 5, "wiek": 52, "dochod": 11000, "wydatki": 6200}
]

# 2. MASZYNKA DO LICZENIA (LAMBDA)
# Bierze klienta (k), kopiuje go (**k) i dokleja mu nową informację: wskaźnik
dodaj_oszczednosci = lambda k: {
    **k, 
    "wskaznik": round((k["dochod"] - k["wydatki"]) / k["dochod"], 2)
}

# 3. URUCHOMIENIE MASZYNKI (MAP)
# Przesyłamy każdego klienta przez maszynkę i robimy nową listę
klienci_z_analiza = list(map(dodaj_oszczednosci, klienci))

# 4. SITKO (FILTER)
# Przepuszczamy tylko tych, którzy oszczędzają 30% (0.30) lub więcej
oszczedni = list(filter(lambda k: k["wskaznik"] >= 0.30, klienci_z_analiza))

# 5. WYŚWIETLANIE WYNIKÓW
print("=== WSZYSCY KLIENCI Z ANALIZĄ ===")
for k in klienci_z_analiza:
    print(f"Klient {k['id']}: Wskaźnik oszczędności = {k['wskaznik']}")

print(f"\nLiczba osób oszczędnych: {len(oszczedni)}")
```
