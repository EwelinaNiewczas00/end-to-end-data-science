Zadanie 9 - Inwersja stownika
Masz słownik mapujacy nazwy produktów na kategorie: ('laptop': 'elektronika' ,
'klawiatura': 'elektronika', 'krzesto': 'meble', 'biurko': 'meble'} . Stwórz
odwrotny stownik gdzie klucze to kategorie, a wartosci to listy produktów.
Wymagania:
• Iteruj po oryginalnym stowniku
• Zgrupuj produkty wedlug kategorii
• Wynik: {'elektronika': ['laptop', 'klawiatura'], 'meble': [...1}
Oczekiwany wynik:
• Stownik kategoria: lista_produktów

```python
słownik = {'laptop': 'elektronika' ,'klawiatura': 'elektronika', 'krzesto': 'meble', 'biurko': 'meble'}
print("Metryki modelu: ")
kategorie ={}

for metryka, wartosc in słownik.items():
    if wartosc in kategorie:
        kategorie[wartosc].append(metryka)
    else:
        kategorie[wartosc] = [metryka]

print(kategorie)
```

Zadanie 11 – Dictionary comprehension - kwadraty liczb
Stwórz słownik używając dictionary comprehension, gdzie klucze to liczby od 1 do 20, a
wartości to ich kwadraty. Następnie przefiltruj ten słownik (też comprehension!) by zawierał
tylko liczby parzyste.
Wymagania:
Użyj dictionary comprehension 2 razy
Pierwszy słownik: wszystkie liczby 1-20
Drugi słownik: tylko parzyste klucze
Dwa słowniki wypisane

```python

słownik = {x: x**2 for x in range(1,20) if x % 2 ==0}
print(słownik)
```

Zadanie 12 – Analiza transakcji - słowniki zagnieżdżone
Masz listę transakcji (krotek): [('2024-01-15', 'Elektronika', 1200), ('2024-01-
16', 'Meble', 800), ('2024-01-15', 'Elektronika', 350), ...]. Stwórz słownik
gdzie kluczem jest data, a wartością słownik z sumą sprzedaży dla każdej kategorii tego
dnia.
Dataset:
transakcje = [
('2024-01-15', 'Elektronika', 1200),
('2024-01-15', 'Meble', 450),
('2024-01-16', 'Elektronika', 800),
('2024-01-16', 'Elektronika', 350),
('2024-01-15', 'Odzież', 200),
('2024-01-16', 'Meble', 600)
]
Wymagania:
Zagnieżdżone słowniki: {data: {kategoria: suma}}
Iteruj po transakcjach i akumuluj sumy
Oczekiwany wynik:
Słownik typu: {'2024-01-15': {'Elektronika': 1200, 'Meble': 450, ...}, ...}

```python
transakcje = [
('2024-01-15', 'Elektronika', 1200),
('2024-01-15', 'Meble', 450),
('2024-01-16', 'Elektronika', 800),
('2024-01-16', 'Elektronika', 350),
('2024-01-15', 'Odzież', 200),
('2024-01-16', 'Meble', 600)
]

wynik = {}

for data, kategoria, kwota in transakcje:
    if data not in wynik:
        wynik[data] = {}
    if kategoria in wynik[data]:
        wynik[data][kategoria] = wynik[data][kategoria] + kwota
    else:
        wynik[data][kategoria] = kwota
print(wynik)
```
