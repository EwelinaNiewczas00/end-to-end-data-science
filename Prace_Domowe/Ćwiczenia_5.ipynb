6. ✏ Zadanie 6 – Formatowanie raportu
Masz dane: model_name = "XGBoost", accuracy = 0.8934, time = 45.67. Użyj f-
string aby stworzyć raport:

Accuracy: 89.34%
Training time: 45.67s

```python
kolumny = ["Model", "Accuracy", "Traning time"]
wiersze = ["XGBoost", "89.34%", "45.67s"]
raport =[]
for kolumny, wiersze in zip(kolumny, wiersze): #zip przypisuje 1 do 1 elementu z listy etc.
    raport.append(f"{kolumny}: {wiersze}")
raport2 = '\n'.join(raport)
print(f"\nRaport:\n{raport2}")
```

9. ✏ Zadanie 9 – Filtrowanie outlierów
Masz listę pomiarów: [23, 24, 22, 150, 23, 25, -10, 24, 23]. Outlierem jest wartość
poza zakresem [20, 30]. Stwórz nową listę bez outlierów używając list comprehension.

```python
pomiary = [23, 24, 22, 150, 23, 25, -10, 24, 23]
lista= []
for wartosc in pomiary:
    if wartosc >= 20 and wartosc <=30:
        lista.append(wartosc)
print(f"Pomiary bez wartości pomiędzy zbiorem zamkniętym 20-30 to: {lista}")
```

10. ✏ Zadanie 10 – Agregacja danych po kategoriach
Masz listę transakcji (kategoria, kwota):
transactions = [
('food', 50),
('transport', 20),
('food', 30),
('entertainment', 40),
('transport', 15),
('food', 25)
]
Oblicz łączną kwotę dla każdej kategorii. Wynik zapisz w słowniku (użyj pętli i warunków).

```python
transactions = [
('food', 50),
('transport', 20),
('food', 30),
('entertainment', 40),
('transport', 15),
('food', 25)
]

słownik = {}

for kategoria, kwota in transactions:
    if kategoria in słownik:
        słownik[kategoria] = słownik[kategoria]+ kwota
    else:
        słownik[kategoria] = kwota


for kategoria, suma in słownik.items():
    print(f"Kategoria: {kategoria:15} | Kwota: {suma} zł") #rezerwacja 15 miejsc dla kategorii

suma_total = sum(słownik.values())
print("-" * 30)
print(f"Łącznie wydano: {suma_total} zł")

#słownik
#.keys - same nazwy
#.values - same liczby
#.items - keys i values 

```
