```python
import numpy as np
import time
import sys

python_list = list(range(10000))
numpy_array = np.array(range(10000))
start = time.time()
doubled_list = [x*2 for x in python_list]
list_time = time.time() - start
print(f"czas dla listy python: {list_time:.4f}s")

start = time.time()
doubled_array = numpy_array *2
numpy_time = time.time() - start
print(f"Czas dla tablicy Numpy: {numpy_time:.4f}s")
speedup = list_time / numpy_time
print(f"\n Numpy jest {speedup:.1f}")


list_size = sys.getsizeof(python_list)
array_size = numpy_array.nbytes
print(f"\nRozmiar listy: {list_size / 1024:.1f} KB")
print(f"Rozmiar tablicy NumPy: {array_size / 1024:.1f} KB")
print(f"Oszczędność pamięci: {(1 - array_size/list_size)*100:.1f}%")
```

### 1. Podstawy i konfiguracja
*   **`import numpy as np`** – standardowa konwencja importowania biblioteki do analizy danych [1].
*   **`np.__version__`** – zwraca aktualnie zainstalowaną wersję biblioteki NumPy [2].
*   **`np.show_config()`** – wyświetla szczegółowe informacje o konfiguracji biblioteki [2].

### 2. Tworzenie tablic
*   **`np.array()`** – tworzy nową tablicę NumPy (tzw. `ndarray`), na przykład ze standardowej listy Pythonowej lub listy zagnieżdżonej [3, 4].
*   **`np.zeros()`** – tworzy tablicę wypełnioną samymi zerami [4].
*   **`np.ones()`** – tworzy tablicę wypełnioną samymi jedynkami [4].
*   **`np.full()`** – tworzy tablicę, w której wszystkie elementy mają tę samą, zdefiniowaną wartość [4].
*   **`np.arange(start, stop, step)`** – generuje sekwencję liczb o określonym kroku, działając podobnie do wbudowanej funkcji `range()` [5].
*   **`np.linspace(start, stop, num)`** – tworzy wektor składający się z podanej liczby równoodległych od siebie wartości w określonym przedziale [5].
*   **`np.eye()`** – tworzy dwuwymiarową macierz identyczności, mającą jedynki na przekątnej [5].

### 3. Atrybuty tablic (właściwości obiektu ndarray)
*   **`arr.shape`** – zwraca kształt tablicy w postaci krotki, informując o liczbie elementów w każdym wymiarze [4].
*   **`arr.ndim`** – zwraca liczbę wymiarów danej tablicy [4].
*   **`arr.dtype`** – pozwala sprawdzić lub określić typ danych elementów tablicy, co bywa kluczowe przy optymalizacji pamięci [3, 6].
*   **`arr.nbytes`** – zwraca rozmiar tablicy w pamięci wyrażony w bajtach [7].

### 4. Modyfikowanie tablic
*   **`arr.reshape()`** – zmienia kształt tablicy bez ingerencji w zawarte w niej dane [6].
*   **`arr.copy()`** – wymusza utworzenie niezależnej w pamięci kopii tablicy, ponieważ standardowe wycinanie zwraca jedynie widok do oryginalnych danych [8].

### 5. Indeksowanie, filtrowanie i slicing
*   **`arr[indeks]`** lub **`arr[wiersz, kolumna]`** – wyciąga konkretny, pojedynczy element z tablicy 1D lub 2D [9, 10].
*   **`arr[start:stop:step]`** – (tzw. slicing) wycina i zwraca wybrany fragment tablicy [9].
*   **`arr[:, kolumna]`** – wycina wszystkie wiersze z wybranej kolumny, co jest użyteczne przy wydobywaniu konkretnych cech [11].
*   **Boolean indexing** – filtruje dane na podstawie maski logicznej, np. `arr[arr > 15]` [12].
*   **Fancy indexing** – przekazuje listę lub tablicę indeksów, aby wyciągnąć precyzyjnie wybrane elementy, np. `arr[[2, 13, 14]]` [15, 16].

### 6. Funkcje matematyczne (ufuncs) i operatory
*   **Operatory arytmetyczne (`+`, `-`, `*`, `/`, `//`, `%`, `**`)** – NumPy obsługuje wektorowo wszystkie klasyczne znaki działań [17].
*   **Operator `@`** – służy do mnożenia macierzowego (dot product) [18, 19].
*   **`np.sqrt()`** – oblicza pierwiastek kwadratowy ze wszystkich elementów tablicy [20].
*   **`np.square()`** – podnosi poszczególne wartości do kwadratu [20].
*   **`np.log()`** – oblicza logarytm naturalny elementów [20].
*   **`np.exp()`** – aplikuje funkcję eksponencjalną do każdego elementu [20].
*   **`np.sin()`, `np.cos()`, `np.pi`** – pozwalają na wyliczanie wartości funkcji trygonometrycznych dla kątów w radianach oraz udostępniają stałą Pi [21].
*   **`np.round()`, `np.floor()`, `np.ceil()`** – to funkcje operujące na zaokrąglaniu wartości (do najbliższej, w dół i w górę) [21].
*   **`np.abs()`** – zwraca wartość bezwzględną z przekazanych danych [21].
*   **`np.clip(tablica, min, max)`** – przycina wartości w tablicy tak, by nie przekraczały podanego przedziału [22].

### 7. Funkcje statystyczne i agregujące
Większość z tych funkcji wspiera argument **`axis`**, operujący wzdłuż kolumn (`axis=0`) lub wierszy (`axis=1`) [23, 24].
*   **`np.mean()`** – oblicza średnią arytmetyczną [22].
*   **`np.median()`** – oblicza medianę rozkładu [22].
*   **`np.std()`** – wylicza odchylenie standardowe ze zbioru [22].
*   **`np.min()`, `np.max()`** – szuka najmniejszej i największej wartości w dostępnej tablicy [22].
*   **`np.sum()`** – sumuje elementy, a przy tablicach logicznych może służyć do zliczania ilości spełnionych warunków [22, 23].
*   **`np.argmin()`, `np.argmax()`** – podają jako wynik indeks, pod którym znajduje się odpowiednio najmniejszy i największy element [25, 26].
*   **`np.percentile()`** – wylicza dany percentyl dla zbioru wartości [23].
*   **`np.unique()`** – zwraca listę unikalnych wartości występujących w datasecie [27].

### 8. Generowanie wartości losowych (moduł `np.random`)
*   **`np.random.rand()`** – generuje losowe wartości należące do przedziału `[0, 1)` [5].
*   **`np.random.randint()`** – tworzy losowe liczby całkowite ze wskazanego przez nas zakresu [5, 26].
*   **`np.random.randn()`** – losuje wartości z rozkładu normalnego [28].
*   **`np.random.seed()`** – definiuje ziarno losowości gwarantujące odtwarzalność wyników za każdym uruchomieniem [29].
*   **`np.random.shuffle()`** – pozwala w miejscu przetasować rozkład elementów w tablicy [30].

### 9. Pozostałe narzędzia
*   **`np.pad()`** – dodaje odpowiedni margines dookoła krawędzi wejściowej tablicy [19, 31].
*   **`np.concatenate()`** – pozwala na elastyczne łączenie różnych tablic ze sobą [30].
*   **`np.argsort()`** – zwraca tablicę indeksów w kolejności potrzebnej do posortowania elementów rosnąco [32].
*   **`np.column_stack()`** – złącza wejściowe tablice jednowymiarowe w pionie, tworząc gotowe kolumny dla nowej tablicy wielowymiarowej [32].
*  **`'Parametr axis jest kluczowy:'`**
*  **`axis=0 : operacja wzdłuż wierszy (kolapsuje wiersze → wynik per kolumna)`**
*  **`axis=1 : operacja wzdłuż kolumn (kolapsuje kolumny → wynik per wiersz)`**
* *  **`'Brak axis : operacja na całej tablicy'`**
