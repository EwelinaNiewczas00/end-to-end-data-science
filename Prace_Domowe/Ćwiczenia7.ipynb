```python
temperatury = [-5, 3, -2, 8, 12, -1, 6, 0, 15, -3]

suma = 0 
pomiar = 0 

for temperatura in temperatury:
    if temperatura > 0:
        suma += temperatura
        pomiar += 1

print(suma)
print(pomiar)
```
