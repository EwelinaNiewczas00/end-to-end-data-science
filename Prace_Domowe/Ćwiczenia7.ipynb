1.65, 1.78, 1.72] (m). Oblicz BMI dla każdego i sklasyfikuj (niedowaga < 18.5, norma
18.5-24.9, nadwaga >= 25).
Dataset:
Listy równoległe: wagi i wzrosty
Wymagania:
Użyj pętli for i in range(len(wagi)):
Oblicz bmi = wagi[i] / (wzrosty[i] ** 2)

Oczekiwany rezultat:

Zadanie 10 – Znajdź wszystkie liczby pierwsze do 100
Napisz program, który znajdzie wszystkie liczby pierwsze od 2 do 100 używając
zagnieżdżonych pętli.
Wymagania:
Pętla zewnętrzna: for n in range(2, 101):
Pętla wewnętrzna: sprawdź dzielniki od 2 do √n
Zmienna czy_pierwsza = True
Jeśli znajdziesz dzielnik, ustaw czy_pierwsza = False i użyj break
Jeśli czy_pierwsza == True, dodaj do listy
Oczekiwany rezultat:
Lista liczb pierwszych: [2, 3, 5, 7, 11, 13, ..., 97]
Liczba znalezionych: 25

```python
import math

pierwsze = []
for n in range(2,101):
    czy_pierwsza = True
    limit = int(math.sqrt(n)) +1
    for i in range(2, limit):
        if n % i ==0:
            czy_pierwsza = False 
            break 
    else:
        pierwsze.append(n)
print(f"Lista liczb pierwszych: {pierwsze}")
print(f"Liczba znalezionych: {len(pierwsze)}")

pierwsze2 = []
for n2 in range(2,101):
    pierwsza = True
    i2 = 2
    while i2 < n2:
        if n2 % i2 ==0:
            pierwsza = False
            break
        i2 += 1
    if pierwsza == True:
        pierwsze2.append(n2)

print(pierwsze2)

 
```

🧠 Zadania wyzwanie (13-20)

Masz dane sprzedaży miesięcznej: [1200, 1350, 1290, 1480, 1520, 1510, 1680,

1720, 1700, 1890, 1950, 2100] . Zidentyfikuj miesiące, w których sprzedaż wzrosła w

porównaniu do poprzedniego miesiąca.

Dataset:

Wymagania:

Oczekiwany rezultat:

(średnie)

Lista 12 wartości (sprzedaż miesięczna w PLN)

Pętla for i in range(1, len(sprzedaz)):

Porównaj sprzedaz[i] z sprzedaz[i-1]

Oblicz zmianę procentową

Wyświetl miesiące ze wzrostem

Lista miesięcy ze wzrostem i procentowy przyrost zrob to

```python
sprzedaz = [1200, 1350, 1290, 1480, 1520, 1510, 1680, 1720, 1700, 1890, 1950, 2100]

for i in range(1, len(sprzedaz)):
    
    teraz = sprzedaz[i]
    wczesniej = sprzedaz[i - 1]
    
    
    if teraz > wczesniej:
        
        zł = teraz - wczesniej
        procent = (zł / wczesniej) * 100
        
        print("Miesiąc numer", i + 1, "miał wzrost o", round(procent, 2), "%")
```
