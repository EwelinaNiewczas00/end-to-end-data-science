Klasa (class) to ogólny projekt, schemat lub wspomniany przepis. Mówi ona komputerowi: "Każdy samochód (Car) będzie miał markę, model i kolor".

Obiekt to konkretny egzemplarz stworzony na podstawie tego schematu. Twoje car1 (Toyota) i my_car (Honda) to właśnie obiekty.

def __init__(self, make, model, color):
To jest tzw. konstruktor. Uruchamia się automatycznie, gdy tworzysz nowy obiekt. Słowo self to sposób, w jaki obiekt mówi o samym sobie (np. "mój kolor to czerwony"). 

self.owner_number = 0
To jest atrybut (cecha). Każdy nowy samochód zaczyna z zerową liczbą właścicieli.

Metody (car_info, sell)
To są "umiejętności" obiektu.

car_info wypisuje dane o samochodzie.

sell zwiększa liczbę właścicieli o 1 przy każdej sprzedaży.
for i in range(5):
    my_car.sell()
    Skoro my_car na początku miał owner_number = 0, a pętla wywołała funkcję sell() dokładnie 5 razy, to jak myślisz, jaką liczbę właścicieli wypisze my_car.car_info() na samym końcu?

```python
class Car(object):
    def __init__(self,make,model,color):
        self.make=make;
        self.model=model;
        self.color=color;
        self.owner_number=0 
    def car_info(self):
        print("make: ",self.make)
        print("model:", self.model)
        print("color:",self.color)
        print("number of owners:",self.owner_number)
    def sell(self):
        self.owner_number=self.owner_number+1
    car1 = Car("Toyota","Corolla","red")
    my_car = Car("Honda","Civic","blue")
    car1.car_info()
    for i in range(5):
        my_car.sell()

    my_car.car_info()
```

```python
class Car(object):
    def __init__(self, make, model, color):
        self.make = make
        self.model = model
        self.color = color
        self.owner_number = 0

    def car_info(self):
        print("make: ", self.make)
        print("model:", self.model)
        print("color:", self.color)
        print("number of owners:", self.owner_number)

    def sell(self, number_of_new_owners):
        self.owner_number = self.owner_number + number_of_new_owners

# Tutaj tworzymy obiekty (już bez wcięcia, bo to główny program)
car1 = Car("Toyota", "Corolla", "red")
my_car = Car("Honda", "Civic", "blue")

car1.car_info()
my_car.sell(5)
my_car.car_info()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Przygotowanie danych (100 punktów od -2 do ok. 6.28)
x = np.linspace(-2, 2 * np.pi, 100)
y = np.sin(x)

# 2. Tworzenie wykresu
plt.figure(figsize=(8, 5))  # Ustawienie rozmiaru okna
plt.plot(x, y, label='sin(x)', color='blue', linewidth=2)

# 3. Dodanie opisów (Dobre praktyki!)
plt.title("Wykres funkcji sinus", fontsize=14)
plt.xlabel("Oś X (radiany)")
plt.ylabel("Wartość sin(x)")
plt.axhline(0, color='black', linewidth=0.5)  # Linia pozioma y=0
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# 4. Wyświetlenie wykresu
plt.show()
```

```python

import time 
import sys
import numpy as np 

import matplotlib.pyplot as plt


def Plotvec1(u, z, v):
    
    ax = plt.axes() # to generate the full window axes
    ax.arrow(0, 0, *u, head_width=0.05, color='r', head_length=0.1)# Add an arrow to the  U Axes with arrow head width 0.05, color red and arrow head length 0.1
    plt.text(*(u + 0.1), 'u')#Adds the text u to the Axes 
    
    ax.arrow(0, 0, *v, head_width=0.05, color='b', head_length=0.1)# Add an arrow to the  v Axes with arrow head width 0.05, color red and arrow head length 0.1
    plt.text(*(v + 0.1), 'v')#Adds the text v to the Axes 
    
    ax.arrow(0, 0, *z, head_width=0.05, head_length=0.1)
    plt.text(*(z + 0.1), 'z')#Adds the text z to the Axes 
    plt.ylim(-2, 2)#set the ylim to bottom(-2), top(2)
    plt.xlim(-2, 2)#set the xlim to left(-2), right(2)
```
