Python calculator 

```python
operator = input("Wpisz operator (+ - * /): ")
war1 = float(input("Wpisz pierwsza liczbe: "))
war2 = float(input("Wpisz druga liczbe: "))

if operator =="+":
    wynik = war1 + war2
    print(round(wynik, 3))
elif operator =="-":
    wynik = war1 - war2
    print(round(wynik, 3))
elif operator =="*":
    wynik = war1 * war2
    print(round(wynik, 3))
elif operator =="/":
    wynik = war1 / war2
    print(round(wynik, 3))
else:
    print(f" Wartoć {operator} jest nieprawidłowy")




```

```python
jednostka = input("Wpisz jednostkę temperatury (C(celcjusz) lub F(Fahrenheit)): ")
zmienna = float(input("Wpisz wartosc temperatury: "))

if jednostka == "f" or "F":
    zmienna = round((zmienna - 32) * 5/9,1)
    print(f"temperatura w Celcjuszach to: {zmienna} °C") 
elif jednostka == "C" or "c":
    zmienna = round((9 * zmienna) /5 + 32, 1) 
    print(f"temperatura w Farenhait to: {zmienna} °F") 
else:
    print(f" Podana jednostka {jednostka} nie jest znana")
```

Indeksowanie  [start : end : step]

```python
#[start: end : step] tzn. np print(zmienna[0 : 2 : 2])
credit_card = "1234-4673-4739-7394"
print(credit_card[4]) #zwraca pozycje ze zeminnej [na 0 indesie(1pozycja)]
print(credit_card[0:4])
print(credit_card[0:4:2]) #co dwa od pozycji 0 do 4
print(credit_card[-2]) #od kona
print(credit_card[::3])
print(credit_card[-4:]) #ostaie 4 cyfry 
print(credit_card[::-1]) #odwraca kod karty od końca)
```

```python
price1 = 3000.14159
price2 = -974.84
price3 = 12.34

print(f"Cena pierwsza to {price1:.2f}")
print(f"Cena pierwsza to {price1:10}") #spacje przed mziennna
print(f"Cena pierwsza to {price1:010}") # 0 przed zmienna 
print(f"Cena pierwsza to {price1:<2}") # warunek do lewej 
print(f"Cena pierwsza to {price1:>22}") #warunek ile do prawej 
print(f"Cena pierwsza to {price1:^10}") # w centrum wydrodkowane
print(f"Cena pierwsza to {price1:+}") #dodane do tekstu +/-
print(f"Cena pierwsza to {price1: }") #dodanie spacji 
print(f"Cena pierwsza to {price1:,}") #zmienne tysiacowe odzielane ,

print(f"Cena pierwsza to {price1:+,.2f}")  #uycie kilku na raz


```
