Python quiz game 

```python

pytania = ("Ile krawedzi ma szecian?: ", "Które zwierzę składa największe jaja?: ", "Jaki jest wynik działania 2+2*2: ", "Jakiego gazu jest najwięcej w atmosferze?: ", "Ile kosci posiada człowiek?: ")
opcje = (("A. 8", "B. 6","C. 16 ","D. 12 "),("A. Słoń ", "B. Krokodyl ","C. Delfin ","D. Strus "),("A. 8 ", "B. 6","C. 12 ","D. 16 "),("A. Tlen ", "B. Dwutlenek węgla ","C. Azot  ","D. Wodów "),("A. 206 ", "B. 207 ","C. 208 ","D. 209 "))
odpowiedzi = ("D" , "D", "B", "C", "A")
osoba = []
punkty = 0
num_pyt = 0



for pytanie in pytania:
    print("--------------")
    print(pytanie) #wypisuje po kolei jedno pytanie 
    for opcja in opcje[num_pyt]: #dopasowuje element tuples pytania do odpowiedzi pozycji
        print(opcja) #wypisuje pozycje opcjii do kazdego pytania
    odp = input("Wpisz (A, B, C, D): ").upper() #kaze wpisac odpowiedx do pytania
    osoba.append(odp) #dodaje odp na koniec listy 
    if odp == odpowiedzi[num_pyt]: #jesli osoba odpowie ta odpowiedzia dobra na pytanie dla kazdego pytania wypisz
        punkty += 1
        print("dobra odpowiedź")
    else:
        print("zła odpowiedź")
        print(f"{odpowiedzi[num_pyt]} to poprawna odpowiedz")

    num_pyt = num_pyt + 1       
print("--------------")
print("   REZULTAT")
print("--------------")

print("Twój wybór: ", end="")
for odp in osoba:
    print(odp, end=" ")
print()

print("Poprawne odpowiedzi: ", end="")
for odp in odpowiedzi:
    print(odp, end=" ")
print()

score = int(punkty / len(pytania) *100)
print(f"Twój wynik to : {score} %")

```
