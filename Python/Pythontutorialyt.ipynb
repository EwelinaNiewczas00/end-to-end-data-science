```python
import math 

a = float(input(" Podaj długosc a: "))
b = float(input("Podaj długosc b: "))
wzór = math.sqrt((pow(a,2))+(pow(b, 2)))

print(f"Bok c (przeciwprostoktna) ma dlugosc: {round(wzór, 2)} cm")
```
