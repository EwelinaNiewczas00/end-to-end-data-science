Prosty kalkulator budzetu

```python
budzet = float(input("Wpisz miesięczny budzet w zł: "))
jedzenie = float(input(" Wpisz kwotę wydawana na jedzenie: "))
rozrywka = float(input("Wpisz kwotę przeznaczona na rozrywkę: "))
rachunki = float(input("Wpisz kwote za opłaty abonanentowe/rachunki(nie pisz wydatków za wynajem) : "))
czynsz = 0
while True:
    wynajem = str(input("Czy wynajmujesz mieszkanie? Napisz TAK jezeli wynajmujesz NIE jezeli nie wynajmujesz (TAK/NIE): "))
    if wynajem.upper()== "TAK":
        czynsz = float(input("Ile płcisz za wynajem: "))
        break
    elif wynajem.upper()== "NIE":
        pass
    else:
        print("Wpisałes niepoprawna odpowiedz musisz wpisac TAK lub NIE")

oszczednosci = float(input("Jaka kwotę odkładasz na czarna godzine/oszczedzasz: "))       
pozostałe = float(input("Wpisz kwotę, na inne pozostałe wydatki, ktore nie zostały ujęte w pytaniach (moze byc 0)w zł: "))


wydatki = jedzenie + rozrywka + rachunki + czynsz + oszczednosci + pozostałe
suma = budzet - wydatki

print("----------------------")
print("PODSUMOWANIE BUDZETU")
print("----------------------")
print(f"W ciagu miesiaca wydajesz {wydatki} zł")
print(f"Twój budzet to: {budzet} zł")
print(f"Zostaje Ci do rozdysponowania {suma}")
```
