Zadanie 1 - Kalkulator BMI

Napisz program, który obliczy wskaźnik BMI (Body Mass Index).

Dane:

• Waga: 70 kg

• Wzrost: 1.75 ml

Wymagania:

• Stwórz zmienne waga i wzrost

• Oblicz BMI według wzoru: BMI = waga / (wzrost')

• Wyświetl wynik z 2 miejscami po przecinku

• Dodaj interpretację: BMI < 18.5 (niedowaga), 18.5-24.9 (norma), ≥25 (nadwaga)

Oczekiwany rezultat:

BMI: 22.86

Interpretacja: Norma#"""

```python
#zadanie1
waga = 70 
wzrost = 1.75 

BMI = waga / (wzrost ** 2)
print (f'dla podanej wagi {waga} kg i wzrostu {wzrost} metra BMI wynosi:' + "  " f'{BMI:.2F}'  , end=" ") 
if BMI < 18.5:
    print("interpretacja: niedowaga")
elif 18.5 <= BMI <= 24.9:
    print("interpretacja: norma")
else:
    print("interpretacja: nadwaga")
```

Określ, który model jest lepszy.Dane:
Wymagania:

Porównaj accuracy (różnica > 0.02 jest znacząca)

```python
#Zadanie 7
accuracy_m1 = 0.87
train_time_m1 = 120
accuracy_m2 = 0.89
train_time_m2 = 450


print(f"model 1 ma lepsze accuracy niz model 2: {accuracy_m1 > accuracy_m2}")
print(f"model 2 ma lepsze accuracy niz model 1: {accuracy_m1 < accuracy_m2}")
print(f"model 1 jest wolniejszy niz model 2: {train_time_m1 > train_time_m2}")
print(f"model 2 jest wolniejszy ni 1: {train_time_m2 > train_time_m1}")

minus = accuracy_m2 - accuracy_m1
if minus <= 0.02:
    print("mała róznica")
elif minus > 0.02:
    print("jest róznica czasowa znaczaca")

minimalny = 0.85
maksymalny = 200
kompromis_m1 = (accuracy_m1 >= minimalny) and (train_time_m1 <= maksymalny)
kompromis_m2 = (accuracy_m2 >= minimalny) and (train_time_m2 <= maksymalny)

if kompromis_m1:
    print("model 1 rekomenduję to najlepszy jest lepszy kompromisem")
elif kompromis_m2:
    print("nodel 2 rekomenduję to najlepszy jest lepszy kompromisem")
else:
    print("nie rekomenduje modeli")
```
